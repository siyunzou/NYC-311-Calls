```python
# importing data via API
import requests
import json
API_311 = requests.get('https://data.cityofnewyork.us/resource/erm2-nwe9.json?$limit=400000')
print(API_311.status_code)
```

```python
df_311 = API_311.text
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sidetable
import geopandas as gpd
from shapely import wkt
```

```python
df = pd.read_json(df_311)
```

```python
# review data and datatype 
df.head(5)
df.dtypes
```

```python
#summary table by agency
df.stb.freq(['agency'])
```

```python
#summary table by borough
df_borough = df.stb.freq(['borough'])
df_borough
```

```python
sns.barplot(data=df_borough, x = "borough", y = "percent")
```

```python
# indexing for year and month
df['year'] = pd.DatetimeIndex(df['created_date']).year
df['month'] = pd.DatetimeIndex(df['created_date']).month
```

```python
df
```

```python
# Review 311 calls by borough
plt.figure(figsize=(10,10))
sns.scatterplot(x='longitude', y ='latitude', hue = 'borough', s=20, data=df)
```

```python
# borough count
borough_ct = df.groupby('borough').agg('count').reset_index()
borough_ct['borough'] = borough_ct['borough'].str.lower()
```

```python
borough_ct
```

```python
# Using geopandas to bring in base layer of NYC boroughs
nyc = gpd.read_file(gpd.datasets.get_path('nybb'))
nyc.head(5)
```

```python
#Rename the column to boroname, so that we can join the data to it on a common field
nyc.rename(columns={'BoroName':'borough'}, inplace=True)
nyc['borough'] = nyc['borough'].str.lower()
bc_geo = nyc.merge(borough_ct, on='borough')
```

```python
bc_geo
```

```python
# plot count by borough into map 
fig,ax = plt.subplots(1,1,figsize=(10,10))
bc_geo.plot(column='unique_key', cmap='viridis_r', alpha=0.5, ax=ax, legend=True)
#bc_geo.apply(lambda x: ax.annotate(s=x.borough, color='black', xy=x.geometry.centroid.coords[0],ha='center'), axis=1)
plt.title("Number of 311 Calls by NYC Borough")
plt.axis("off")
```
